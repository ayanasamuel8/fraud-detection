## Part 1: Data Cleaning & Correction

### 1. Correcting Time-Based Data
**Hypothesis (If X...):** If we can measure the time between events, we can uncover behavioral patterns unique to fraudsters. The `signup_time` and `purchase_time` columns are currently stored as text (object), preventing any time-based calculations.

**Action (Then Y...):** Then we must convert these columns to a datetime format.

**Result (Therefore Z...):** Therefore, we unlock the ability to engineer crucial, time-sensitive features, such as the time between signup and purchase.

```python
import pandas as pd
import numpy as np

# Convert to datetime
fraud_df['purchase_time'] = pd.to_datetime(fraud_df['purchase_time'])
fraud_df['signup_time'] = pd.to_datetime(fraud_df['signup_time'])

# Verify the change
fraud_df.info()
```

Now that the time-based columns are in the correct format, we can engineer a new feature: the time difference between signup and purchase. This can be a powerful indicator of fraudulent activity.

```python
fraud_df['time_to_purchase'] = (fraud_df['purchase_time'] - fraud_df['signup_time']).dt.total_seconds()

# Display the new feature
fraud_df[['signup_time', 'purchase_time', 'time_to_purchase', 'class']].head()
```

Let's visualize the distribution of `time_to_purchase` for both fraudulent and legitimate transactions to see if there are any discernible patterns.

```python
import plotly.express as px

fig = px.histogram(
    fraud_df,
    x='time_to_purchase',
    color='class',
    nbins=100,
    log_y=True,
    title='Distribution of Time to Purchase by Class',
    labels={'time_to_purchase': 'Time from Signup to Purchase (seconds)'}
)
fig.show()
```

**Visual Analysis Summary:**

The histogram reveals a distinct pattern. A significant number of fraudulent transactions occur very quickly after signup, often within a day. This suggests that `time_to_purchase` is a valuable feature for our fraud detection model.

### 2. Standardizing IP Addresses
**Hypothesis (If X...):** If we can map a transaction's IP address to a country, we can identify high-risk geographic regions. The `ip_address` is currently a float, which could lead to precision errors and is incompatible for merging with the IP-to-country mapping table.

**Action (Then Y...):** Then we must convert the `ip_address` column to a 64-bit integer type, which is a standard, precise format for IP numbers.

**Result (Therefore Z...):** Therefore, the data is now prepared for a fast and accurate merge with the `IpAddress_to_Country.csv` dataset, enabling geolocation analysis.

```python
# Convert ip_address to int64
fraud_df['ip_address'] = fraud_df['ip_address'].astype(np.int64)

# Verify the change
fraud_df.info()
```

With the `ip_address` standardized, we can now enrich our data by mapping IP addresses to their respective countries. Let's load the IP-to-country mapping data and merge it with our fraud dataset.

```python
# Load IP to Country data
ip_to_country_df = load_data('../data/00_raw/IpAddress_to_Country.csv')
ip_to_country_df.head()
```

```python
# Merge dataframes to get country information
def map_ip_to_country(ip):
    try:
        return ip_to_country_df[(ip_to_country_df['lower_bound_ip_address'] <= ip) & (ip_to_country_df['upper_bound_ip_address'] >= ip)]['country'].iloc[0]
    except IndexError:
        return 'Unknown'

fraud_df['country'] = fraud_df['ip_address'].apply(map_ip_to_country)

fraud_df[['ip_address', 'country', 'class']].head()
```

### A More Robust Approach: Bayesian Averaging for Fraud Rates

The challenge with picking a `min_transactions` threshold is that it's arbitrary. A better way to handle this is with **Bayesian Averaging**.

This method adjusts a country's calculated fraud rate by "shrinking" it towards the global average fraud rate. The amount of shrinkage depends on the number of transactions:
- **Countries with many transactions:** Their calculated rate is trusted, so it changes very little.
- **Countries with few transactions:** Their rate is unreliable, so it gets pulled more strongly towards the global average.

This gives us a more stable and reliable estimate of risk for each country.

```python
# Calculate the global fraud rate
global_fraud_rate = fraud_df['class'].mean()

# Set a "credibility" parameter (C). This is like adding C "ghost" transactions with the global average rate.
# A common choice for C is the average number of transactions per country.
C = fraud_rate_df['total_transactions'].mean()

# Calculate the Bayesian-adjusted fraud rate
fraud_rate_df['bayesian_fraud_rate'] = (
    (fraud_rate_df['total_transactions'] / (fraud_rate_df['total_transactions'] + C)) * (fraud_rate_df['fraudulent_transactions'] / fraud_rate_df['total_transactions']) +
    (C / (fraud_rate_df['total_transactions'] + C)) * global_fraud_rate
) * 100 # Convert to percentage

# Sort by the new adjusted rate
top_bayesian_rate_countries = fraud_rate_df.sort_values(by='bayesian_fraud_rate', ascending=False).head(15)

print("Top 15 Countries by Bayesian-Adjusted Fraud Rate:")
print(top_bayesian_rate_countries[['total_transactions', 'fraudulent_transactions', 'fraud_rate', 'bayesian_fraud_rate']])
```

```python
# Visualize the top 15 countries by Bayesian-adjusted fraud rate
fig = px.bar(top_bayesian_rate_countries, 
             x=top_bayesian_rate_countries.index, 
             y='bayesian_fraud_rate',
             title='Top 15 Countries by Bayesian-Adjusted Fraud Rate',
             labels={'x': 'Country', 'y': 'Adjusted Fraud Rate (%)'},
             hover_data=['total_transactions', 'fraudulent_transactions', 'fraud_rate'])
fig.show()
```

**Bayesian Averaging Analysis Summary:**

This chart provides our most reliable view of country-level risk. Notice how some countries that had very high raw fraud rates (but low transaction counts) are no longer at the top. The countries that remain have a high fraud rate that is supported by a reasonable number of transactions.

This adjusted fraud rate will be a much more robust feature for our machine learning model.


## Part 2: Final Feature Engineering

### 1. Engineering Time-Based Patterns
**Hypothesis (If X...):** If fraudsters operate at specific times to avoid detection (e.g., late at night), then the hour of purchase will be a predictive signal.

**Action (Then Y...):** Then we will extract the hour of the day and the day of the week from the `purchase_time` column.

**Result (Therefore Z...):** Therefore, we create features that allow the model to learn time-based behavioral patterns.

```python
# Extract hour of day and day of week
fraud_df['purchase_hour'] = fraud_df['purchase_time'].dt.hour
fraud_df['purchase_day_of_week'] = fraud_df['purchase_time'].dt.dayofweek # Monday=0, Sunday=6

fraud_df[['purchase_time', 'purchase_hour', 'purchase_day_of_week']].head()
```

```python
# Visualize fraud counts by hour of day
fig = px.histogram(fraud_df, 
                   x='purchase_hour', 
                   color='class',
                   barmode='group',
                   title='Fraud vs. Legit Transactions by Hour of Day',
                   labels={'purchase_hour': 'Hour of Day'})
fig.show()
```

### 2. Engineering Device-Based Anomaly Detection
**Hypothesis (If X...):** If a single physical device is used to access many different user accounts, it is highly indicative of a coordinated fraud ring.

**Action (Then Y...):** Then we will create a feature that counts how many unique users are associated with each `device_id`.

**Result (Therefore Z...):** Therefore, we create a feature that directly quantifies device sharing, a powerful signal for detecting organized fraud.

```python
# Calculate number of unique users per device
device_user_counts = fraud_df.groupby('device_id')['user_id'].nunique().to_dict()
fraud_df['device_user_count'] = fraud_df['device_id'].map(device_user_counts)

fraud_df[['device_id', 'user_id', 'device_user_count', 'class']].sort_values('device_user_count', ascending=False).head(10)
```

```python
# Visualize the distribution of device_user_count for fraudulent transactions
fig = px.histogram(fraud_df[fraud_df['class']==1], 
                   x='device_user_count',
                   title='Distribution of Device Sharing for Fraudulent Transactions',
                   labels={'device_user_count': 'Number of Users per Device'})
fig.show()
```

## Part 3: Finalizing the Dataset for Modeling
Our feature engineering is now complete. The final step before modeling is to clean up the dataframe, keeping only the features we will use.

```python
# Add the Bayesian-adjusted fraud rate to the dataframe
fraud_rate_df.rename(columns={'bayesian_fraud_rate': 'country_risk_rate'}, inplace=True)
fraud_df = fraud_df.merge(fraud_rate_df['country_risk_rate'], on='country', how='left')

# Drop identifier/redundant columns
final_df = fraud_df.drop(columns=[
    'user_id', 
    'signup_time', 
    'purchase_time', 
    'device_id', 
    'ip_address',
    'country' # Drop original country column after merging risk rate
])

# One-hot encode categorical variables
final_df = pd.get_dummies(final_df, columns=['source', 'browser', 'sex'], drop_first=False)

print("Final dataset columns:")
print(final_df.columns)
final_df.head()
```

```python
final_df.drop_duplicates(inplace=True)
```

```python
final_df = final_df.loc[:, ~final_df.columns.duplicated(keep='first')]
```
